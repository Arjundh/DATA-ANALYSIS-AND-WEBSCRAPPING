```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support.expected_conditions import  visibility_of_element_located
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import re
import pandas as pd
import numpy as np
import time
```

```python
all_section=[]
driver = webdriver.Firefox(executable_path='C://Users//Arjun D H//geckodriver-v0.24.0-win64//geckodriver.exe')
for index in range(0,5):
    print(index)
    url = "https://www.naukri.com/walkin-jobs"
    driver.get(url)
    driver.implicitly_wait(50)
    section = driver.find_elements_by_xpath("//*[@class='row  ']")
    for i in section:
        all_section.append(i.text)
driver.close()
```

```python
print(len(all_section))
```

```python
print(all_section[0])
```

```python
import pandas as pd
df=pd.DataFrame(index=range(0,250),columns=['job_Role','company','reviews','experience','location','description'])
for idx,section in enumerate(all_section):
    scn=section.split('\n')
    try:
        df.iloc[idx,0]=scn[0] #job_role
    except IndexError:
        df.iloc[idx,0]=np.nan
    try:
        df.iloc[idx,1]=scn[1] #company
    except IndexError:
        df.iloc[idx,1]=np.nan
    try:
        df.iloc[idx,2]=scn[2]#reviews
    except IndexError:
        df.iloc[idx,2]=np.nan
    try:
        df.iloc[idx,3]=scn[3] #experience
    except IndexError:
        df.iloc[idx,3]=np.nan
    try:
        df.iloc[idx,4]=scn[4] #location
    except IndexError:
        df.iloc[idx,4]=np.nan
    try:
        df.iloc[idx,5]=scn[5] #description
    except IndexError:
        df.iloc[idx,5]=np.nan
```
